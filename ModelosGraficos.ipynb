**Projeto** 1: Elabore em programa em Python que faça o gráfico da amplitude do oscilador harmônico amortecido no intervalo de tempo $t=0$ a $t=30$. O coeficiente de amortecimento é $\gamma =0.1$ e a frequência natural de oscilação é $\omega=2.0$. Dado que:

$f=A e^{-\gamma t}cos(\omega t)$

```python
#cap3_graf1v1.py
'''Exemplo de grafico simples com matplotlib
importando todos os metodos
Programador: Elinei Santos
Data da ultima revisao: 26/12/2016'''

from matplotlib.pylab import * #importa todas as funções
from math import *
t = linspace(0,30,200) #criar uma lista em python tipo um vetor
y = zeros(len(t)) #len busca o tamanho de um vetor
for i in range(len(t)):
  f = 8*exp(-0.1*t[i])*cos(2*t[i])
  y[i] = f
plot(t,y,'-b') #plota os dois vetores de uma vez y versus t
title('Oscilador amortecido')
xlabel('tempo(s)', fontsize = 15)
ylabel('y(m)', fontsize = 15)
axis([0,30, -8,8])
show()
```

```python
#cap3_graf1v2.py
'''Exemplo de grafico simples com legenda
usando o modulo matplotlib
Programador: Elinei Santos
Data da ultima revisao: 26/12/2016'''

from matplotlib.pylab import *
from math import *
t = linspace(0,30,200)
y1 = zeros(len(t))
y2 = zeros(len(t))
for i in arange(len(t)):
    f1 = 8*exp(-0.2*t[i])*cos(2.*t[i])
    f2 = 10*exp(-0.15*t[i])*cos(2.5*t[i])
    y1[i] = f1
    y2[i] = f2
plot(t,y1,'bD',linewidth = 2)
#hold('on') #utilizado em versoes mais antigas
plot(t,y2,'r-', linewidth = 3)
title('Oscilador harmonico amortecido')
xlabel('tempo(s)', fontsize = 15)
ylabel('y(m)', fontsize = 15)
legend(['figura1', 'figura2'])
axis([0,30, -10,10])
show()
```

```python
#cap3_graf4.py
'''Exemplo de grafico simples com matplotlib
importando todos os metodos
Programador: Elinei Santos
Data da ultima revisao: 26/12/2016'''

from matplotlib.pylab import *
from math import *
t = linspace(0,30,200)
y1 = zeros(len(t))
y2 = zeros(len(t))
y3 = zeros(len(t))
for i in arange(len(t)):
    f1 = 4*exp(-0.1*t[i])*cos(1.2*t[i]+pi/6)
    f2 = 4*exp(-0.2*t[i])+2*exp(-0.2*t[i])
    f3 = (2.+3*t[i])*exp(-0.3*t[i])
    y1[i] = f1
    y2[i] = f2
    y3[i] = f3
plot(t,y1,'bo')
plot(t,y2,'k^')
plot(t,y3,'rs')
xlabel('tempo(s)', fontsize = 15)
ylabel('y(m)', fontsize = 15)
legend(['f1-azul', 'f2-preto', 'f3-vermelho'],loc=1)
axis([0,30, -6,6])
show()
```

```python
#trajetoria2_obliquo.py
'''Programa para visualizar a trajetoria de um projétil
Programador: Elinei Santos
Data da ultima revisao: 12/04/2023'''

from matplotlib.pylab import *
from math import *
g=9.8

#calcular o tempo de voo do projetil
vo = float(input('Entre com a velocidade inicial de lançamento :\n'))
te=float(input('Entre com o angulo de lancamento em graus: \n'))
teta=(te*pi)/180 #converte para radianos
t_voo=2.*vo*sin(teta)/g
tmin=0.
tmax=t_voo
t=[]

h=0.01
tempo=tmin
while tempo < tmax:
    t.append(tempo)
    tempo=tempo+h

x1=zeros(len(t))
y1=zeros(len(t))

for i in range(len(t)):
    x = vo*cos(teta)*t[i]
    y = vo*sin(teta)*t[i]-0.5*g*t[i]*t[i]
    x1[i]=x
    y1[i]=y

plot(x1,y1,'-b',linewidth=2)
title('Trajetória - lançamento obliquo')
xlabel('x(m)', fontsize = 15)
ylabel('y(m)', fontsize = 15)
#axis([0,30, -8,8])
show()
```

Exemplo: conjunto de dados Iris do Scikit-Learn, em que cada amostra é um dos três tipos de flores com tamanho de sua pétalas e sépalas medido.


```python
from sklearn.datasets import load_iris
iris = load_iris()
'''load_iris() retorna um dicionário-like (Bunch) com:
data: matriz (150, 4) com medidas de pétalas e sépalas.
target: vetor com as classes (0 = setosa, 1 = versicolor, 2 = virginica).
feature_names: nomes das 4 variáveis (sepal length, sepal width, petal length, petal width)'''
features = iris.data.T
'''iris.data normalmente é (150, 4) → cada linha = flor, cada coluna = característica.
.T transpõe a matriz → (4, 150), ou seja:
features[0] = vetor com o comprimento da sépala para todas as flores.
features[1] = vetor com a largura da sépala.
features[2] = vetor com o comprimento da pétala.
features[3] = vetor com a largura da pétala.   '''
plt.scatter(features[0], features[1], alpha=0.5, s=100*features[3], c = iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
